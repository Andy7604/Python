## Funciones

## Funciones que llaman otras funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}!"

def despedirse(nombre):
  return f"Adios, {nombre}!"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return saludo + "\n" + despedida

print(mensaje_completo("Andrea"))
```

```python
# Ahora supongamos que tenemos una heladeria
def pedir_sabor():
  sabores = ["vainilla","chocolate","fresa"]
  print("Sabor disponible", ",", sabores)
  # lower() hace minusculas las palabras
  # strip() elimina los espacios tanto antes como después
  sabor = input("Elige un sabor: ").strip().lower()

  while sabor not in sabores:
    sabor = input("Elige nuevamente un sabor: ").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("Cuantas bolas de nieve quieres?: "))
  while cantidad <= 0:
    cantidad = int(input("Cuantas bolas de nieve quieres?, ingresa un valor válido: "))
  return cantidad

# Función madre (llama a otras funciones)
def pedir_helado():
  sabor = pedir_sabor()
  cant = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de {sabor}")
  return cant, sabor

total_pedidos = int(input("Cuantos clientes atenderas?: "))
for i in range(total_pedidos):
  pedir_helado()
```

## Funciones Lambda
Una función lambda es una función acortada para escribir una función de esta forma:

**nombreFuncion = lambda parametros: acción**

La diferencia entre las funciones lambda y las funciones normales es que suelen ser un poco más limitadas.

```python
sumar = lambda x,y : x + y
print(sumar(2,3))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1-desc)), (precio * desc))
nuevoPrecio = descuentoLambda(0.4, 500)
print(type(nuevoPrecio))
print("El precio a pagar es $", nuevoPrecio[0], ", el descuento que se hizo fue de: ", nuevoPrecio[1])
```

```python
# Calculadora -> suma, resta, producto, división y potencia
def suma(a,b):
  return a + b
def resta(a,b):
  return a - b
def producto(a,b):
  return a * b
def division(a,b):
  if b == 0:
    return "Error: no se puede dividir ente 0"
  return a / b
def potencia(a,b):
  return a**b


def menu():
  print("------ MI CALCULADORA ------")
  print("1. Suma")
  print("2. Resta")
  print("3. Producto")
  print("4. División")
  print("5. Potencia")
  print("6. Salir")

while True:
  menu()
  try:
    op = int(input("Seleccione una opción: "))
    if op < 1 or op > 6:
      print("Opción invalida")
  except ValueError:
    print("Ingrese un número válido")
  if op == 6:
    print("SALIENDO...")
    break

  try:
    a = float(input("Ingrese el primer número: "))
    b = float(input("Ingrese el segundo número: "))
  except ValueError:
    print("Error: ingrese un número válido")
  if op == 1:
    print("Resultado: ", suma(a,b))
  elif op == 2:
    print("Resultado: ", resta(a,b))
  elif op == 3:
    print("Resultado: ", producto(a,b))
  elif op == 4:
    print("Resultado: ", division(a,b))
  elif op == 5:
    print("Resultado: ", potencia(a,b))
  else:
    print("Opción no válida")
```
