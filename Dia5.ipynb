# Existe 1 solo tema que TODO programador independientemente del area DEBE de saber
JSON es un formato para guardar y transmitir datos. Es muy usado en APIs, archivos de  de configuración y almacenamiento de información.
{Se parece a un diccionario de Python, pero es solo texto}

¿Para qué sirve?
- Sirve para convertir objetos de python -> texto JSON y tambiém texto JSON -> objetos de Python

Esto se llama:
- Serializar: Convertir un objeto de Python en texto JSON.
- Deserializar: Convertir un texto JSON a un objeto usable de Python.

```python
# Funciones principales:
# 1) json.dums(obj): Convierte un diccionario u objeto de Python en un string JSON
import json
obj = {"nombre" : "Andrea",
       "edad" : 21}
print(type(obj))
s = json.dumps(obj, ensure_ascii = False)
print(s)
print(type(s))
```

```python
# json.loads(texto_json): Convierte un string json a un diccionario Python
restaurar = json.loads(s)
print(restaurar)
print(type(restaurar))
# Al venir de una página/persona/archivo desconocido debes de tener cuidado
Va = '''
Pueden contener:
- Datos malinformados
- Información inesperada
- Campos faltantes
- Datos demasiado grandes
- Variables incorrectas que rompan tú programa6
'''
print(Va)
```

## Numpy
Provee el tipo de array, operaciones vectorizadas, broadcasting, utilidades de algebra lineal y funciones matemáticas en arrays. Es la base en la que se apoyan muchas librerias científicas en Python.

### Nota IMPORTANTE: El arreglo de numpy no es igual a la lista de Python, este NO puede tener diferentes tipos de datos en el mismo.

```python
#pip install numpy
import numpy as np
A = np.array([[1,2,3],[4,5,6]]) # array (2,3)
v = np.array([10,20,30]) #array (3)
print(A + v)
print("---------------------")
prueba = [1,2,3]
prueba2 = [1,2,3]
print(prueba + prueba2)
```

```python
# Más funciones de numpy
z = np.zeros((3,3)) #array (3,3)
print(z)
print("---------------------")
o = np.ones((2,4))
print(o)
print("---------------------")
r = np.random.rand(3,3)
print(r)
```

```python
# Rangos numéricos
nums = np.arange(0,10,2) # obtiene los valores entre 0 y 10 de dos en dos sin incluir el 10
print(nums)
x = np.linspace(0,1,5) # obtiene 5 valores espaciados de manera equitativa entre 0 y 1
print(x)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
print(a + b) #suma
print(a - b) #resta
print(a * b) #multiplicación
print(a / b) #división
print(a ** b) #potencia
```

```python
print(a.dot(b)) #producto entre dos matrices
c = np.array([[1,2,3],[4,5,6]])
d = np.array([[10,20],[30,40],[50,60]])
print(c.dot(d))
```

```python
data = np.array([1,2,3])
print(data.min()) #mínimo
print(data.max()) #máximo
print(data.sum()) #suma
print(data.mean()) #promedio
print(data.std()) #desviación estandar
```

```python
A = np.array([1,6,3,9,2])
print(A[A > 5])
```
