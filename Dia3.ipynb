# Estructura de datos
Las estructuras de datos son formas de organizar informaci√≥n para trabajar con ella de manera eficiente.

- Listas -> ordenadas, modificadas, permiten duplicados
- Tuplas -> ordenadas, inmutables, permiten duplicados
- Sets -> no ordenados, sin duplicados
- Diccionarios -> pares clave:valor

Nota:
- Pilas (FILO = First In Last Out)
- Colas (FIFO = First In First Out)
- Listas enlazadas (o doblemente enlazadas)

```python
#OPCIONAL
import random
opciones = ["piedra", "papel", "tijeras"]

while True:
  while True:
    eleccion = input("\nElige piedra, papel o tijeras (o escribe salir si ya no quieres jugar): ")
    if eleccion not in opciones and not eleccion == "salir":
      print("Opci√≥n no v√°lida, intenta de nuevo.")
    else:
      break

  if eleccion == "salir":
    print("¬°Gracias por jugar!")
    break
  aleatoria = random.choice(opciones)
  print("La computadora eligi√≥: ", aleatoria)

  #Determinamos el ganador
  if eleccion == aleatoria:
    print("Empate")
  elif (eleccion == "piedra" and aleatoria == "tijeras") \
    or (eleccion == "papel" and aleatoria == "piedra" ) \
    or (eleccion == "tijeras" and aleatoria == "papel"):
   print("¬°Ganaste!")
  else:
    print("Perdiste")
```

```python
# Uni√≥n
matutino = {20322,22183,12345,76453}
vespertino = {20322,1111,22222,33333}
# Union con .union
total_matriculas = matutino.union(vespertino)
print(total_matriculas)
# Union pero con |
total_matriculas = matutino | vespertino
print(total_matriculas)
```

```python
# Intersecci√≥n
ca = {1234, 121, 646, 123443, 73456, 1223243,678192}
al = {1234, 111,222,333,444}
# intersecci√≥n con .intersection()
ca_al = ca.intersection(al)
print(ca_al)
# interseccion usando &
ca_al = ca & al
print(ca_al)
```

```python
# Diferencia
# Diferencia usando .difference
solo_calculo = ca.difference(al)
print(solo_calculo)
# Diferencia usando -
solo_calculo = ca -al
print(solo_calculo)
```

## Diccionarios
Son colecciones no ordenadas de pares clave:valor

```python
alumno = {
    #clave       valor
    "nombre" : "Andrea Carrillo",
    "edad" : 21,
    "carrera" : "ITS",
    "semestre": 7,
    "promedio" : 92,
    "calificaciones" : [100,94,94,97,100]
}
print(type(alumno))
alumno
```

```python
# ========================================
# TIENDA SIMPLE - Versi√≥n optimizada
# ========================================

# Lista de productos (ordenada)
productos = ["manzana", "banana", "leche", "pan"]

# Tupla de precios (no se modifican)
precios = (10, 5, 20, 15)

# Diccionario de inventario: producto ‚Üí {"precio": x, "stock": y}
inventario = {
    productos[i]: {"precio": precios[i], "stock": 10}
    for i in range(len(productos))
}

# Set de productos en oferta (b√∫squeda r√°pida O(1))
ofertas = {"manzana", "pan"}

# ========================================
# Mostrar cat√°logo de productos
# ========================================
print("üçé BIENVENIDO A LA TIENDA üçû\n")
print("‚ïê" * 50)
for p in productos:
    info = inventario[p]
    mensaje = "üî• ¬°OFERTA ESPECIAL! üî•" if p in ofertas else "üí∞ Precio regular"
    print(f"{p.title():8} ‚Üí ${info['precio']:2} | Stock: {info['stock']:2} ‚Üí {mensaje}")
print("‚ïê" * 50)

# ========================================
# Sistema de compra interactivo
# ========================================
print("\nüõí ¬øQu√© producto deseas comprar?")
compra = input("Escribe el nombre del producto: ").strip().lower()

# Buscar si el producto existe (independiente de may√∫sculas)
producto_encontrado = None
for p in productos:
    if p.lower() == compra:
        producto_encontrado = p
        break

if producto_encontrado and inventario[producto_encontrado]["stock"] > 0:
    inventario[producto_encontrado]["stock"] -= 1
    precio = inventario[producto_encontrado]["precio"]
    oferta_texto = " (¬°EN OFERTA!)" if producto_encontrado in ofertas else ""
    print(f"\n‚úÖ ¬°Compra exitosa!{oferta_texto}")
    print(f"   Compraste: 1 {producto_encontrado.title()}")
    print(f"   Precio: ${precio}")
    print(f"   Stock restante: {inventario[producto_encontrado]['stock']}")
else:
    print("\n‚ùå Lo siento, el producto no est√° disponible o no tenemos stock.")

# ========================================
# Mostrar inventario actualizado (opcional)
# ========================================
print("\nüì¶ Inventario actualizado:")
for p in productos:
    info = inventario[p]
    print(f"   ‚Ä¢ {p.title():8} ‚Üí Stock: {info['stock']}")
```
